Requirements:

1.	How does the distribution of patient ages vary by gender?

```python
sns.boxplot(x='Gender', y='Age', data=health_data)#Using seaborn
plt.title('Distribution of Patient Ages by Gender')
plt.show()
summary_stats = health_data.groupby('Gender')['Age'].describe()
print(summary_stats)
```

2.	How does the average billing amount differ by medical condition?

```python
health_data['Billing Amount'] = pd.to_numeric(health_data['Billing Amount'], errors='coerce')
avg_billing = health_data.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending=False)
print(avg_billing)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_billing.index, y=avg_billing)
plt.title('Average Billing Amount with respect to Medical Condition')
plt.ylabel('Average Billing Amount')
plt.xlabel('Medical Condition')
plt.show()
```

3.	What is the distribution of blood types among patients?

```python
blood_type_counts = health_data['Blood Type'].value_counts()
print(blood_type_counts)
plt.figure(figsize=(8, 8))
plt.pie(blood_type_counts, labels=blood_type_counts.index, autopct='%1.1f%%')
plt.title('Blood Type Distribution')
plt.show()
```

4.	How does the length of stay (discharge date - admission date) vary by admission type?

```python
health_data['Date of Admission'] = pd.to_datetime(health_data['Date of Admission'])
health_data['Discharge Date'] = pd.to_datetime(health_data['Discharge Date'])
health_data['Length of Stay'] = (health_data['Discharge Date'] - health_data['Date of Admission']).dt.days
print(health_data['Length of Stay'])
plt.figure(figsize=(10, 6))
sns.violinplot(x='Admission Type', y='Length of Stay', data=health_data)
plt.title('Length of Stay Distribution variation by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Length of Stay')
plt.show()
```

13.	What is the relationship between blood type and medical condition?

```python
contingency_table = pd.crosstab(health_data['Blood Type'], health_data['Medical Condition'])
print(contingency_table)
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d')
plt.title('Relationship Between Blood Type and Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Blood Type')
plt.show()
```
